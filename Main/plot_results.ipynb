```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("BenchResults.csv")
data.head()
```

```python
# Set the style of seaborn
sns.set(style="whitegrid")

# Create a line plot
plt.figure(figsize=(10, 6))
sns.lineplot(x='Task', y='Execution Time (ms)', hue='Tool', style = 'Tool', data=data, markers=['o', 'X'], palette='Set1')

# Set plot labels and title
plt.xticks(rotation=0, ha='center', fontsize=7)
plt.xlabel('Task')
plt.ylabel('Execution Time (ms)')
plt.title('Execution Time for Tasks by Tool')

# Save the plot to a file
plt.savefig('Bench_plots/Execution_time.png')
# Show the plot
plt.show()

```
